```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('Data/carprices.csv')
df.head(3)
```

```python
plt.title('Graph of BMW Price')
plt.xlabel('Sell Price in US$')
plt.ylabel('Mileage of the car')
plt.scatter(df['Mileage'], df['Sell Price($)'])
```

```python
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
plt.xlabel('Sell Price in US$')
plt.ylabel('Age of the car in year')
plt.title('the Graph of BMW year and Sell Price')
```

```python
X = df[['Mileage', 'Age(yrs)']]
X
```

```python
y = df['Sell Price($)']
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

By train_test_split method we can split our data in to train and test catagories but normally it chooses randomly if you want to choose constantly use random_state = 10 inside the trai_test _split mothod as a parameter

```python
print('Size of X_train:' + str(len(X_train)))
print(f'Size of X_test:' + str(len(X_test)))
print('size of all dataset:' + str(len(X)))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
model.predict(X_test)
```

```python
print(y_test)
```

now if we want to check the accuracy of our model we can use score method it compare the Y_test with the data of predict(X_test)

```python
model.score(X_test,y_test)
```
